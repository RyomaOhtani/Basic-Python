## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv("/Users/fuka/Desktop/Basic-Python/data/hist_data.csv")
plt.hist(df["aデータ"], bins = 30, label = "aデータ", color = "#f0c311", alpha = 0.5)
plt.hist(df["bデータ"], bins = 30, label = "bデータ", color= "c", alpha = 0.3)
plt.legend(loc="upper right", fontsize=10)
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.title("aデータとbデータの温度に関する度数分布", fontsize = 15)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv("/Users/fuka/Desktop/Basic-Python/data/scatter_data.csv")
class0_df = df[df["class"] == 0]
class1_df = df[df["class"] == 1]
class2_df = df[df["class"] == 2]
plt.scatter(class0_df["length (cm)"], class0_df["width (cm)"], c = "r", label = "class0")
plt.scatter(class1_df["length (cm)"], class1_df["width (cm)"], c = "b", label = "class1")
plt.scatter(class2_df["length (cm)"], class2_df["width (cm)"], c = "g", label = "class2")
plt.grid(True)
plt.title("クラス0~2のlengthとwidthに関する散布図")
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
plt.legend(loc="upper right")
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv("/Users/fuka/Desktop/Basic-Python/data/line_data.csv")
plt.plot(df["期間"], df["件数"], marker = "o")
plt.grid(True)
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")
plt.xticks(rotation=30)
plt.show()

plt.bar(df["期間"] ,df["費用"])
plt.grid(True)
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.title("期間ごとの費用の推移")
plt.xticks(rotation=30)
plt.show()
```
