```python
df = pd.read_csv("data/line_data.csv")
plt.plot(df["期間"], df["件数"], marker = "o")
plt.grid(True)
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")
plt.xticks(rotation=30)
plt.show()

plt.bar(df["期間"] ,df["費用"])
plt.grid(True)
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.title("期間ごとの費用の推移")
plt.xticks(rotation=30)
plt.show()
```

```python

number_periods_costs = pd.read_csv("data/line_data.csv")
fig = plt.figure(
    figsize=(6, 9)
)
ax1 = fig.add_subplot(
    211,
    title="期間ごとの件数の推移",
    xlabel="期間",
    ylabel="件数",
    )
ax2 = fig.add_subplot(
    212,
    title="期間ごとの費用の推移",
    xlabel="期間",
    ylabel="費用(円)",
    )
ax1.grid(True)
ax1.set_xticklabels(number_periods_costs["期間"], rotation=30, ha="right")
ax1.plot(number_periods_costs["期間"],number_periods_costs["件数"], marker="o")

ax2.set_xticklabels(number_periods_costs["期間"], rotation=30, ha="right")
ax2.grid(True)
ax2.bar(number_periods_costs["期間"], number_periods_costs["費用"])

fig.tight_layout()
```

```python
import matplotlib.pyplot as plt

df_line_data = pd.read_csv("data/line_data.csv")
def configure_plot(ax, title, x_label, y_label):
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.grid(True)
    ax.tick_params(axis='x', labelrotation=30)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# 折れ線グラフの作成
ax1.plot(df_line_data["期間"], df_line_data["件数"], marker="o")
configure_plot(ax1, "期間ごとの件数の推移", "期間", "件数")

# 棒グラフの作成
ax2.bar(df_line_data["期間"], df_line_data["費用"])
configure_plot(ax2, "期間ごとの費用の推移", "期間", "費用(円)")

plt.tight_layout()
plt.show()
```
